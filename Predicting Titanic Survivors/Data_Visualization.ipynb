```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv("titanic_data_modelready.csv")
df.head()
```

```python
# survied vs did not survied
survival_counts= df['Survived'].value_counts()
plt.figure()
plt.bar(['Did not survived', 'survived'] , survival_counts)
plt.xlabel('Survival status')
plt.ylabel('Number of passengers')
plt.title('Survival distribution')
plt.savefig("survival_distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Survival by gender 
gender_survival = df.groupby(['Sex' , 'Survived']).size().unstack()
gender_survival.columns = ['MALE','FEMALE']
plt.figure()
gender_survival.plot(kind="bar")
plt.xlabel("Gender")
plt.ylabel('Number of passengers ')
plt.title('survived by Gender')
plt.xticks(rotation=0)
plt.savefig("gender_survival.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure()
plt.hist(df['Age'].dropna(), bins=30)
plt.xlabel("Age")
plt.ylabel("Number of passengers ")
plt.title('Age Distribution of Passengers')
plt.savefig("AGE_Distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```
